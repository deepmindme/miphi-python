Для начала обсудим условия решения задачи:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import redef extract_experience_months(text):    if pd.isna(text):        return np.nan    text_str = str(text)    if text_str == 'Не указано':        return np.nan    match = re.search(r'Опыт работы\s+(\d+)\s+(?:лет|года|год)', text_str)    years = 0    months = 0    if match:        years = int(match.group(1))    match_months = re.search(r'(\d+)\s+месяц(?:а|ев)?', text_str)    if match_months:        months = int(match_months.group(1))    if years == 0 and months == 0:        return np.nan    return years * 12 + months
```

```python
if 'Опыт работы' in df.columns:    df['Опыт работы (месяц)'] = df['Опыт работы'].apply(extract_experience_months)    df['Опыт работы (месяц)'].dtype    df.drop(columns=['Опыт работы'], inplace=True)median_experience = df['Опыт работы (месяц)'].median()print(f'Медианный опыт работы: {median_experience:.0f} месяцев')
```

**Комментарий к графику:**

1. **Хвост распределения тянется вправо** - в сторону больших значений возраста
2. **Медиана меньше среднего** - большинство соискателей имеют возраст меньше среднего значения
3. **Наличие выбросов справа** - есть небольшое количество соискателей с очень высоким возрастом, которые значительно превышают типичные значения

Такая асимметрия характерна для распределения возраста в контексте поиска работы, так как:
- Большинство активных соискателей находятся в возрасте от 25 до 50 лет
- Количество соискателей старшего возраста постепенно уменьшается
- Наличие резюме людей преклонного возраста (выбросы) создает длинный правый хвост распределения

Использование логарифмического масштаба помогает лучше визуализировать такую асимметрию и более точно определить выбросы методом z-отклонений.
